# < Pie  >

```python
import matplotlib.pyplot as plt
```

```python
# default :counter-clockwise


my_labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
my_sizes = [15, 30, 45, 10]
my_explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

plt.figure()
plt.pie(my_sizes, explode=my_explode, labels=my_labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0.0, 20.0, 0.01)
s = np.sin(t)
plt.plot(t, s, 'm-.', label = 'sine')

plt.xlabel('time (s)')
plt.ylabel('voltage (mV)')
plt.title('Graphs of Sine')
plt.grid(True)
plt.legend(loc = 9, bbox_to_anchor = (0.9, -0.1))
plt.savefig("test_graph.png")
plt.show()
```
