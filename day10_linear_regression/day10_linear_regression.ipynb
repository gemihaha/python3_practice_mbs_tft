### Train/Test Set 분할

보통은 실제 데이터를 바탕으로 Train을 하고, 새로운 데이터를 기반으로 Test를 하지만, 본 연습에서는 주어진 데이터를 Train 데이터와 Test 데이터로 나누어 분석하고자 한다. 이를 위해 Scikit-Learn이 제공하는 train_test_split 모듈을 사용한다.

```python
from sklearn.model_selection import train_test_split

features = ['Weather', 'Weekend', 'Baseball', 'Temp']
X_train, X_test, y_train, y_test = train_test_split(df[features], df['Sales'], random_state = 2)                #75:25로 랜던 분할(default)
#X_train, X_test, y_train, y_test = train_test_split(df[features], df['Sales'], test_size=0.15)#85:15로 랜덤 분할
```

# Linear Regression

#### Linear regression 수행

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from matplotlib import pyplot as plt

```

```python
lm = linear_model.LinearRegression()
lm.fit(X_train, y_train)
```

#### 통계값 계산

```python
# The coefficients
print('Coefficients: {}'.format(lm.coef_))
# The mean squared error
print("Mean squared error: {}".format(mean_squared_error(y_train, lm.predict(X_train))))
# Explained variance score: 1 is perfect prediction
print('R^2: {}'.format(r2_score(y_train, lm.predict(X_train))))
# Accuracy
print("Training set score: {}".format(lm.score(X_train, y_train)))
print("Test set score: {}".format(lm.score(X_test, y_test)))
```

#### 전체 판매량 재예측 및 시각화

학습된 lm 모델을 이용하여 전체 데이터를 다시 predict

```python
X = df[features]
X
```

```python
pred = lm.predict(X)
pred
```

```python
df['Predict'] = pred
df
```

```python
df.plot(y=['Sales', 'Predict'], figsize=(15, 10), linewidth=5)
```

# Logistic Regression

데이터가 두 집단(0 또는 1)으로 나뉘어져 있는 경우




분류(classification)에 해당




diagnosis = 'M' : malignant 악성종양

diagnosis = 'B: : benign    양성종양

```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import numpy as np 

df=pd.read_csv('data/diagnosis_data.csv')
df.head()
```

```python
sns.countplot(x='diagnosis', data=df)
plt.show()
```

- 로지스틱 회귀분석은 두 집단이 0 또는 1로 이루어져야 하기 때문에, 

  M과 B 라는 두 집단을 0 과 1로 대체

```python
df['diagnosis']=df['diagnosis'].replace('M',0)
df['diagnosis']=df['diagnosis'].replace('B',1)

sns.countplot(x='diagnosis', data=df)
plt.show()
```

- 모델을 만들 training set과 모델을 평가 할 tset set을 7:3 비율로 나누기

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

Y=df['diagnosis']
X=df.iloc[:,2:32]
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state = 3)
```

로지스틱 회귀분석을 적용하고 평가

```python
log_clf = LogisticRegression(solver = 'liblinear')
log_clf.fit(X_train,Y_train)
log_clf.score(X_test, Y_test)

```

-> 약 95%의 정확도로 M과 B 두그룹을 분류할 수 있는 분류기(classifier)
