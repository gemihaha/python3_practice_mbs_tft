### Train/Test Set 분할

보통은 실제 데이터를 바탕으로 Train을 하고, 새로운 데이터를 기반으로 Test를 하지만, 본 연습에서는 주어진 데이터를 Train 데이터와 Test 데이터로 나누어 분석하고자 한다. 이를 위해 Scikit-Learn이 제공하는 train_test_split 모듈을 사용한다.

```python
from sklearn.model_selection import train_test_split

features = ['TV', 'radio', 'newspaper']
X_train, X_test, y_train, y_test = train_test_split(df[features], df['sales'])                #75:25로 랜던 분할(default)
#X_train, X_test, y_train, y_test = train_test_split(df[features], df['Sales'], test_size=0.15)#85:15로 랜덤 분할
```

### Linear regression

#### Linear regression 수행

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
from matplotlib import pyplot as plt
```

```python
lm = linear_model.LinearRegression()
model = lm.fit(X_train, y_train)
```

#### 통계값 계산

```python
# The coefficients
print('Coefficients: {}'.format(lm.coef_))
# The mean squared error
print("Mean squared error: {}".format(mean_squared_error(y_train, lm.predict(X_train))))
# Explained variance score: 1 is perfect prediction
print('Variance score: {}'.format(r2_score(y_train, lm.predict(X_train))))
# Accuracy
print("Training set score: {}".format(lm.score(X_train, y_train)))
print("Test set score: {}".format(lm.score(X_test, y_test)))
```

#### 전체 판매량 재예측 및 시각화

학습된 lm 모델을 이용하여 전체 데이터를 다시 predict

```python
X = df[features]
X
```

```python
pred = lm.predict(X)
pred
```

```python
df['Predict'] = pred
df
```

# Logistic Regression

당뇨병이 있는지 없는지 알아보자!

< Columns >
- Pregnancies : Number of times pregnant
- Glucose : Plasma glucose concentration a 2 hours in an oral glucose tolerance test
- BloodPressure : Diastolic blood pressure (mm Hg)
- SkinThickness : Triceps skin fold thickness (mm)
- Insulin : 2-Hour serum insulin (mu U/ml)
- BMI : Body mass index (weight in kg/(height in m)^2)
- DiabetesPedigreeFunction : Diabetes pedigree function
- Age : Age (years)

```python
#import pandas
import pandas as pd

# load dataset
dia = pd.read_csv("diabetes.csv")
```

```python
dia.head()
```

< Selecting Feature >

Outcome에 해당하는 열의 데이터만 y에 넣어주고 

나머지 열 이름은 feature_cols 변수에 할당해서 x에 넣어주기

```python
#split dataset in features and target variable
feature_cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin','BMI', 'DiabetesPedigreeFunction', 'Age']
x = dia[feature_cols] # Features
y = dia.Outcome       # Target variable
```

```python
x
```

```python
y
```

x와 y를 training 과 testing sets으로 나누기 size = 0.25

```python
# split X and y into training and testing sets

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

로지스틱 회귀분석을 적용

```python
# import the class
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression(solver= 'liblinear')

# fit the model with data
logreg.fit(x_train,y_train)
```

## .score() 로 확인!
score()안에는 (test samples, true labels for x) 형태가 들어가야 함.

```python
logreg.score(x_test, y_test)
```
