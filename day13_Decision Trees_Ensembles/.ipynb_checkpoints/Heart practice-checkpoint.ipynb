# Heart

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Data Load

```python
df1 = pd.read_csv('data/heart.csv')
df2 = pd.read_csv('data/heart2.csv')

print(df1.head(10))
print(df2.head(10))
```

## pre-processing

#### heart2.csv ID 처리

```python
for i in range(len(df2['id'])) : 
    df2['id'][i] = df2['id_2'][i] + 10000
    
df2.head(20)
```

#### Merge DataFrame

```python
mg_heart = pd.merge(df1, df2, on='id')
mg_heart.head(10)
```

#### thalach : null값 채우기(대표값 아무거나 사용 (ex.mean,median…)

```python
mg_heart[mg_heart['thalach'].isnull()]
```

```python
mean_thalach = mg_heart.loc[mg_heart['thalach'].notnull(), 'thalach'].mean()
mean_thalach
```

```python
mg_heart['thalach'].fillna(mean_thalach, inplace=True)
```

#### cp : “three”는 3으로 코딩해야겠지?

```python
mg_heart['cp'].unique()
```

```python
mg_heart.loc[mg_heart['cp'].apply(lambda x : not x.isdigit()), 'cp'] = 3
```

```python
mg_heart['cp'].unique()
```

#### age : null값이 있는 column은 다 drop!

```python
mg_heart[mg_heart['age'].isnull()]
```

```python
#mg_heart = mg_heart[np.isfinite(mg_heart['age'])]

# subset 을 안해주면 제대로 drop이 안됨
mg_heart.dropna(subset=['age'], inplace=True)
```

```python
mg_heart[mg_heart['age'].isnull()]
```

```python
mg_heart.head(100)
```

#### exang : null값은 다 0을 의미

```python
mg_heart[mg_heart['exang'].isnull()]
```

```python
mg_heart['exang'].fillna(0.0, inplace=True)
```

```python
mg_heart[mg_heart['exang'].isnull()]
```
