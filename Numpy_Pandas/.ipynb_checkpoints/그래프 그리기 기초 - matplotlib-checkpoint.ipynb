# 그래프 그리기 기초 - matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.figure() # 그래프를 그리겠다.
x = [1,3,5]
y = [2,3,6]
plt.plot(x,y) # 선형 그래프를 그리는 method
plt.show() # 그래프 그리기 !
```

```python
a = plt.figure()
a = plt.plot([1,2,3,4,5,6,7,8,9,8,7,6,5,4,3,2,1,0])

b = plt.figure()
b = plt.plot([1,2,3,4,5,6,7,8,9,8,7,6,5,4,3,2,1,0] , marker = 'o', linestyle ='--')

plt.show(a)
plt.show(b)
```

```python
import numpy as np

t = np.arange(0,12, 0.01)  # 0~12 까지 0.01 의 간격으로 
print(t)
y = np.sin(t)
```

```python
plt.figure(figsize = (10,6))
plt.plot(t,y)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(t,y)
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(t,y)
plt.grid()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')
plt.show()
```

```python
plt.figure(figsize =(10,6))
plt.plot(t, np.sin(t))
plt.plot(t, np.cos(t))
plt.grid()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')

plt.show()
```

```python
plt.figure(figsize =(10,6))
plt.plot(t, np.sin(t), label='sin')
plt.plot(t, np.cos(t), label='cos')
plt.grid()
plt.legend()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')

plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(t, np.sin(t), lw=10 , label='sin')
plt.plot(t, np.cos(t), 'r', label='cos')
plt.grid()
plt.legend()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(t, np.sin(t), lw=3 , label='sin')
plt.plot(t, np.cos(t), 'r', label='cos')
plt.grid()
plt.legend()
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')
plt.ylim(-1.2, 1.2)
plt.xlim(0, np.pi)
plt.show()
```

```python
t = np.arange(0, 5, 0.5)

plt.figure(figsize=(10,6))
plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'b+:')
plt.plot(t, t**3, 'g^')
plt.show()
```

```python
t = np.arange(0, 5, 0.5)

plt.figure(figsize=(10,6))
plt1 = plt.plot(t, t**2, 'bs')

plt.figure(figsize=(10,6))
plt2 = plt.plot(t, t**3, 'g^')

plt.show()
```

```python
t = [0,1,2,3,4,5,6]
y = [1,4,5,8,9,5,3]

plt.figure(figsize=(10,6))
plt.plot(t, y, color='green')
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.plot(t, y, 'g--')
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.plot(t, y, color = 'green', linestyle='dashed', marker = 'o',
        markerfacecolor = 'blue', markeredgecolor = 'red', markersize = 15)

plt.xlim([-0.5, 6.5])
plt.ylim([0.5, 9.5])
plt.show()
```

```python
s1 = np.random.normal(loc=0, scale=1, size=1000)
s2 = np.random.normal(loc=5, scale=0.5, size=1000)
s3 = np.random.normal(loc=10, scale=2, size=1000)

print(s1)
len(s1)
```

```python
plt.figure(figsize=(10,6))
plt.plot(s1, label='s1')
plt.plot(s2, label='s2')
plt.plot(s3, label='s3')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.boxplot((s1,s2,s3))
plt.grid()
plt.show()
```

```python
t = np.arange(0, 5, 0.01)

plt.figure(figsize=(10,12))

plt.subplot(411)
plt.plot(t, np.sqrt(t))
plt.grid()

plt.subplot(423)
plt.plot(t, t**2)
plt.grid()

plt.subplot(424)
plt.plot(t, t**3)
plt.grid()

plt.subplot(413)
plt.plot(t, np.sin(t))
plt.grid()

plt.subplot(414)
plt.plot(t, np.cos(t))
plt.grid()

plt.show()
```
