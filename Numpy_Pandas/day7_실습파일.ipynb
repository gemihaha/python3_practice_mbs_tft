#### practice 7-4: 다음의 array는 선택된 20개 나라의 남성과 여성의 school completion data이다. 남녀의 completion 차가 가장 큰 나라를 찾는 함수를 작성하시오

````python
### My solution
def diff_max() : 
    tmp_arr = []
    max_idx = 0 
    for i in range(len(countries)) : 
        tmp_arr.append(abs(female_completion[i] - male_completion[i]))
    

    max_idx = tmp_arr.index(max(tmp_arr))
    print(countries[max_idx])

    

#print(tmp_arr)
#print(tmp_arr.index(max(tmp_arr)))
#print(countries[1])

### Teacher's solution

def diff_completion(countries, female_completion, male_completion) : 
    diff = []
    for i in range(len(countries)) : 
        diff.append(abs(female_completion[i] - male_completion[i]))
    diff = np.array(diff)
    
    return diff.max()
````

#### practice 7-5
* 주어진 employment.csv파일은 header가 없는 파일이다. 불러온 후 column의 header를 추가한다.
* 국가명을 인덱스로 지정한다.
* 불필요한 컬럼이 있으면 삭제한다.
* 상위 5개를 출력한다.

````python
data = pd.read_csv('data/employment.csv')

data2 = data.copy()

data2.columns = ["Nation","employment"]
data2.head()

data2.index = data2.Nation
data2.head()
data2 = data2.drop(['Nation'], axis = 1)
data2.head()
````

#### practice 7-6
* 다음 2개의 파일도 불러와서 employment.csv와 같은 방식으로 편집한다.
* life_expectancy.csv
* gdp_per_capita.csv

````python
# life_expectancy.csv
data = pd.read_csv('data/life_expectancy.csv')
data.head()
data3 = data.copy()
data3
data3.columns = ["Nation", "life_expect"]
data3.head()

data3.index = data3.Nation
data3.head()
data3 = data3.drop(['Nation'], axis = 1)
data3.head()
#data3.index = data3.Nation
#data3.head()
````

````python
# gpd_per_capita.csv
data = pd.read_csv('data/gdp_per_capita.csv', header =None)
data.head()
data4 = data.copy()
data4.head()
data4.columns = ["Nation","GDP"]
data4.head()
data4.index = data4.Nation
data4.head()
data4 = data4.drop(["Nation"], axis = 1)
data4.head()
````

#### practice 7-7
* 세개의 데이터프레임을 하나로 합치자.
* 데이터프레임을 합칠 때는 merge, join 등의 개념이 사용된다.
* http://pandas.pydata.org/pandas-docs/stable/merging.html 를 참고하자.
* (힌트) concat 이 사용된다.

````python
result = pd.concat([data2,data3,data4], axis = 1, join = 'inner')
result.head()
````

소팅을 해보자. 소팅은 다음과 같이 한다.
* data.sort_values([colname1, colname2, ...], ascending=[True, False, ...])

````python
result.sort_values(['employment','life_expect','GDP'], ascending=True)
````
