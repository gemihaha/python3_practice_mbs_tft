# 그래프 그리기 기초 - matplotlib

```python
import matplotlib.pyplot as plt
```

```python
plt.figure() # 그래프를 그리겠다.
x = [1,3,5]
y = [2,3,6]
plt.plot(x,y) # 선형 그래프를 그리는 method
plt.show() # 그래프 그리기 !
```

```python
a = plt.figure()
a = plt.plot([1,2,3,4,5,6,7,8,9,8,7,6,5,4,3,2,1,0])

b = plt.figure()
b = plt.plot([1,2,3,4,5,6,7,8,9,8,7,6,5,4,3,2,1,0] , marker = 'o', linestyle ='--')

plt.show(a)
plt.show(b)
```

```python
import numpy as np

t = np.arange(0,12, 0.01)  # 0~12 까지 0.01 의 간격으로 
print(t)
y = np.sin(t)
```

```python
plt.figure(figsize = (10,6))
plt.plot(t,y)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(t,y)
plt.grid()
plt.show()
```
